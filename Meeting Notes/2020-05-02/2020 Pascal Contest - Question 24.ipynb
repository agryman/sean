```python
def check_sean(x, y):
    z = x * y
    a = x - y
    b = (x + y) / 2
    return b ** 2 == z + a
```

```python
check_sean(9, 5)
```

```python
check_sean(100, 50)
```

```python
check_sean(100, 51)
```

```python
check_sean(100, 96)
```

```python
checks = [(y + 4, y, check_sean(y + 4, y)) for y in range(100)]
```
